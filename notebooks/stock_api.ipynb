```python
from ib_async import IB, Stock, util
util.startLoop()  # uncomment this line when in a notebook
# Connect to IB Gateway (port 7497 for paper trading, ???? for live trading)
ib = IB()
ib.connect('127.0.0.1', 7497, clientId=1)

print("Connected to IB Gateway")
```

```python
#Create a Portfolio
portfolio = [
    Stock('IDVY', 'BVME', 'EUR'),
    Stock('IUKD', 'LSE', 'GBP'),
    Stock('IGCB', 'LSE', 'GBP'),
    Stock('SYBC', 'IBIS2', 'EUR'),
    Stock('PST', 'BVME', 'EUR'),
    Stock('ISP', 'BVME', 'EUR'),
    Stock('NVDA', 'NASDAQ', 'USD'),
    Stock('MSFT', 'NASDAQ', 'USD'),
    Stock('TSM', 'NYSE', 'USD'),    # ADR
    Stock('ENI', 'BVME', 'EUR'),
    Stock('ENEL', 'BVME', 'EUR'),
    Stock('ENGI', 'SBF', 'EUR'),
]
```

```python
for s in portfolio:

    ticker = s.symbol
    print(s)
    contract = s
    bars = ib.reqHistoricalData(
        contract, endDateTime='', durationStr='30 D',
        barSizeSetting='1 Day', whatToShow='MIDPOINT', useRTH=True, formatDate=1)

    # convert to pandas dataframe :
    df = util.df(bars)
    if df is not None:
        df.to_csv(ticker + '.csv', index=False)
    else:
        print(f"No data for {ticker}, skipping CSV export.")
    # print(df)
```

```python
import yfinance as yf
import pandas as pd

# list of tickers, e.g. portfolio
portfolio = ["AAPL", "MSFT", "GOOG", "ENEL.MI", "ENGI.PA"]

for ticker in portfolio:
    print(f"Downloading data for {ticker}...")
    
    # download last 30 days of daily data (similar to 30 D, 1 Day bars)
    df = yf.download(ticker, period="30d", interval="1d", auto_adjust=False)
    
    if not df.empty:
        # reset index for cleaner CSV export
        df.reset_index(inplace=True)
        # export to CSV
        csv_name = f"{ticker.replace('.', '_')}.csv"
        df.to_csv(csv_name, index=False)
        print(f"Saved {csv_name}")
    else:
        print(f"No data for {ticker}, skipping CSV export.")
```

```python
import matplotlib.pyplot as plt
ticker = 'AAPL'
contract = Stock(ticker, 'SMART', 'USD')

# Function to get historical data
def get_historical_data():
    bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='1 Y',
        barSizeSetting='1 Day',
        whatToShow='MIDPOINT',
        useRTH=True,
        formatDate=1
    )
    df = util.df(bars)
    return df
df = get_historical_data()

# Compute Moving Averages
df['SMA_9'] = df['close'].rolling(9).mean()
df['SMA_21'] = df['close'].rolling(21).mean()
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label= ticker + ' Price', color='blue', linewidth=1)
plt.plot(df['date'], df['SMA_9'], label='SMA_9', color='red', linestyle='dashed')
plt.plot(df['date'], df['SMA_21'], label='SMA_21', color='green', linestyle='dotted')

plt.xlabel("Date")
plt.ylabel("Price")
plt.title(ticker + " Price with SMA_9 & SMA_21")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
print("close is: ",df['close'].iloc[-1])
print("SMA 9 is: ",df['SMA_9'].iloc[-1])
print("SMA 21 is: ",df['SMA_21'].iloc[-1])
```
